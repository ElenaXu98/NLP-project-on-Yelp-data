## data preprocessing

```python
import pandas as pd
import numpy as np
```

### tip_city.json and user_city.json

```python
tip = pd.read_json('data/tip_city.json',lines=True)
tip = tip[['business_id','text','user_id']]
tip = tip.dropna()
tip.head()
```

```python
tip.shape
```

```python
user = pd.read_json('data/user_city.json',lines=True)
user = user[['average_stars','fans','review_count','useful','user_id','yelping_since']]
user = user.dropna()
user.head()
```

```python
user.shape
```

### review_city.json

```python
review = pd.read_json('data/review_city.json',lines = True)
review = review[review['useful']==1] # only useful reviews
# review = review[review['date']] # only review in recent one year
review = review[['business_id','review_id','stars','text','user_id']]
review = review.dropna()
review.head()
```

```python
review.columns
review.shape
```

```python
# merge with user
review_user = pd.merge(review,user,how="left",on="user_id")
review_user.head()
```

```python
review_user.shape
```

```python
type(review_user.text)
```

```python
# delete reviews that have exactly the same text and deal with the user
# suspicious to be fake reviews
review = review.drop_duplicates('text')
review.shape
```

### business_city.json

```python
business = pd.read_json('data/business_city.json',lines=True)
business = business.dropna()  # drop missing entries
business = business[business.is_open==1]  # business with is.open == 1
business = business[['business_id', 'categories', 'city', 'hours','name','review_count', 'stars', 'state','attributes']]
business.head()
```

```python
business.shape
```

```python
type(business)
```

```python
index = list()
for bus in business.business_id:
    i = business[business.business_id==bus]
    dic = i.attributes.array[0]
    #print(type(dic))
    #print(dic.keys())
    #print(type(dic.keys()))
    if 'Alcohol'in dic.keys():
        alcohol = dic['Alcohol']
        #print(alcohol)
        #print(type(alcohol))
        if "full_bar" in alcohol:  # only business with alcohol == 'full_bar'
            index.append(True)
        else:
            index.append(False)
    else:
        index.append(False)       
#print(index)
len(index)
```

```python
business = business[index]
```
